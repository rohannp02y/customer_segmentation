**Goal:** Group mall customers into distinct segments based on their **Annual Income** and **Spending Score**, so businesses can target each group with personalized marketing strategies.

**Dataset:** Mall Customers — 200 customers with age, gender, annual income, and spending score  
**Algorithm:** K-Means Clustering  
**Key Technique:** Elbow Method to find the optimal number of clusters

## 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

## 2. Data Collection and Analysis

```python
#Data Collection and Analysis
customer_data = pd.read_csv('../data/Mall_Customers.csv')
customer_data.head()
```

```python
#finding the number of rows and columns
customer_data.shape
#getting info about the datasets
customer_data.info()
```

```python
customer_data.isnull().sum()
```

## 3. Selecting Features

```python
X = customer_data.iloc[:,[3,4]].values
print(X)
```

## 4. Finding the Optimal Number of Clusters (Elbow Method)

**WCSS** = Within Cluster Sum of Squares

```python
#finding wcss value for different number of clusters
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
#plot an elbow graph

sns.set()
plt.plot(range(1,11), wcss)
plt.title('The elbow point Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## 5. Training the K-Means Model

Optimum number of clusters = **5** (from elbow graph)

```python
kmeans= KMeans(n_clusters=5,init='k-means++',random_state=0)
#return a label for each data ppoints based on their table
Y = kmeans.fit_predict(X)
print(Y)
```

## 6. Visualising the Clusters

```python
#plotting al the clusters and their cendroids
plt.figure(figsize=(8,8))
plt.scatter(X[Y==0,0], X[Y==0,1], s=50, c='green', label='Clustern 1')
plt.scatter(X[Y==1,0], X[Y==1,1], s=50, c='red', label='Clustern 2')
plt.scatter(X[Y==2,0], X[Y==2,1], s=50, c='yellow', label='Clustern 3')
plt.scatter(X[Y==3,0], X[Y==3,1], s=50, c='violet', label='Clustern 4')
plt.scatter(X[Y==4,0], X[Y==4,1], s=50, c='blue', label='Clustern 5')

#plot the centroids
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100,c ='cyan',label='Centroids' )
plt.title('Customer Clusters')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
